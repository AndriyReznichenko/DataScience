```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
def calcY(x):
    return (.0006 * x**5) + (.01 * x**4) + (0.2 * x**3)+(-2 * x**2)+(3 * x)+1
```

```python
xs = np.linspace(-5, 10, 201)
ys = np.array([calcY(x) for x in xs])
```

```python
plt.plot(xs, ys)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(input_shape = (None, 1), 
                          units = 80, 
                          activation = lambda x: tf.nn.leaky_relu(x, alpha = 0.4)),
    tf.keras.layers.Dense(units = 40, 
                          activation = lambda x: tf.nn.leaky_relu(x, alpha = 0.4)),
    tf.keras.layers.Dense(units = 20, 
                          activation = lambda x: tf.nn.leaky_relu(x, alpha = 0.4)),
    tf.keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer='adam',
              loss = 'mse',
              metrics=['mse', 'mae'])
```

```python
xs = xs.reshape(-1, 1)
ys = ys.reshape(-1, 1)
```

```python
model.fit(xs, ys, epochs = 250)
```

```python
preds = model.predict(xs)
```

```python
plt_x = xs.reshape(-1)
plt_y = ys.reshape(-1)
plt_pred = preds.reshape(-1)


plt.plot(plt_x, plt_y, color = 'green')
plt.plot(plt_x, plt_pred, color = 'blue')
```
